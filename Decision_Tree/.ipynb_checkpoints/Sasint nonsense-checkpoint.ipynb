```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
trainingDataset = open("training.txt").readlines()
testingDataset = open("testing.txt").readlines()
```

```python
training_labels = []
trn_attri = [[] for i in range(0,128)]
print(len(trainingDataset))
for i in trainingDataset:
#     print(i)
    spl = i.split(" ")
    label = int(spl[0])
    del spl[0]
    for d in range(0,128):
        n = spl[d].split(":")
        trn_attri[int(n[0])].append(int(n[1]))
    training_labels.append(label)
#     print("i: "+label)
trn = {i:trn_attri[i] for i in range(0,128)}
trn["label"] = tuple(training_labels)
```

```python
trn_attri
```

```python
df = pd.DataFrame(trn)
df.head()
```

```python
df.shape
```

```python
X = df.drop("label",axis=1)
y = df["label"]
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
test_attri = [[] for i in range(0,128)]
for i in testingDataset:
#     print(i)
    spl = i.split(" ")
    for d in range(0,128):
        n = spl[d].split(":")
        test_attri[int(n[0])].append(int(n[1]))
#     print("i: "+label)
test = {i:test_attri[i] for i in range(0,128)}
```

```python
df2 = pd.DataFrame(test)
```

```python
df2.head()
```

```python
pred = dt.predict(df2)
```

```python
print(len(pred))
```

```python
result = open("results.txt","w")
lines = []
for i in pred:
    lines.append(str(i)+"\n")
lines.append("---")
print(lines)
result.writelines(lines)
```
